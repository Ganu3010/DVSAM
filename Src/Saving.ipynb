```python
import random
ide, local, spon, nov, roo5, charg, cuis, wnb, vo, vego, sfo, buff=[],[],[],[],[],[],[],[],[],[],[],[]
userid = []
```

```python
for i in range(20000):
    user_id = random.randint(0, 100)
    userid.append(user_id)
    item_id = random.randint(0, 12188)
    ide.append(f2['index'][item_id])
    local.append(f2['Locality'][item_id])
    spon.append(f2['Sponsored'][item_id])
    roo5.append(f2['Ratings_out_of_5'][item_id])
    nov.append(f2['Number of votes'][item_id])
    charg.append(f2['Charges_for_two'][item_id])
    cuis.append(f2['Cuisines'][item_id])
    wnb.append(f2['Wine and Beer'][item_id])
    vo.append(f2['Vegetarian Only'][item_id])
    vego.append(f2['Vegan Options'][item_id])
    sfo.append(f2['Sugar Free Options'][item_id])
    buff.append(f2['Buffet'][item_id])
newdf = pd.DataFrame()
newdf['user_id'] = userid
newdf['index'] = ide
newdf['loc'] = local
newdf['sponsor'] = spon
newdf['votes'] = nov
newdf['rating'] = roo5
newdf['charge'] = charg
newdf['cuisine'] = cuis
newdf['alcohol'] = wnb
newdf['veg'] = vo
newdf['vegan'] = vego
newdf['sugarfree'] = sfo
newdf['buff'] = buff
```
