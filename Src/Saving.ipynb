```python
def get_restaurant(user_dict):
    sim_rest = pd.DataFrame()
    m = np.array(list(user_dict.values())).mean()
    for rest in user_dict:
        sim_rest = sim_rest.append(get_sim(rest, user_dict[rest], m), ignore_index = True)
    sim_rest = sim_rest.sum().sort_values(ascending = False)
    return dict(sim_rest.head(len(user_dict)+5))
```
